# Stacks
A stack is a data structure that is often linked to a stack. If you have just a washed a plate, you put it on the top of the stack. When you need a plate, ypu take it off the top of the stack. So the last palte to be added to the stack will be the first to be removed from stack. Thus the stack is a **last in, first out(LIFO)** structure:


There are two primary operations that are done on stacks: **push** and **pop.** When an element
is added to the top of the stack, it is pushed onto the stack. When an element is taken off the
top of the stack, it is popped off the stack. Another operation which is used sometimes is
peek, which makes it possible to see the element on the stack without popping it off.

```python
# Implementation
class Node:
    def __init__(self, data = None):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        self.size = 0
        
    def push(self, data):
        '''The push operation used to add an element to the top of te stack'''
        node = Node(data)
        if self.top:
            node.next=self.top
            self.top = node
#             print(self.data.top)
        else:
            self.top=node
        self.size+=1
        return data
    def pop(self):
        '''The pop operation used to removed top element from stack. we need to return  the topmost element as well.
        we will return make the return None if there is no more elements:'''
        if self.top:
            data = self.top.data
            self.size-=1
            if self.top.next:
                self.top = self.top.next
            else:
                self.top = None
                
            return data
        else:
            return None
        
    def peek(self):
        '''The Peek will just return the top of the stack without removing it from the stack. '''
        if self.top:
            return self.top.data
        else:
            return None
            
```

```python
st = Stack()
st.push(20)
```
